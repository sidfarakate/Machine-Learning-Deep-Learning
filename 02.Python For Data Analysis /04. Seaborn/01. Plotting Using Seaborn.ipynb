# 1. Relational Plots

- Bivariate Analysis
- To see the statistical relation between 2 or more variables.

#### Plots under this section:
- Scatterplot
- Lineplot

```python
tips = sns.load_dataset('tips')
tips.sample(2)
```

## 1.1. Scatterplot

```python
# scatterplot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# scatterplot using relplot -> figure level function -> square shape
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

## 1.2. Lineplot

```python
gap = px.data.gapminder()
gap.sample(1)
```

```python
temp_df = gap[gap['country'] == 'India']
```

```python
# lineplot -> axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# lineplot using relplot -> figure level function
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
temp = gap[gap['country'].isin(['India', 'Argentina', 'Portugal'])]
```

```python
sns.relplot(data=temp, x='year', y='lifeExp', hue='country', kind='line')
```

```python
sns.relplot(data=temp, x='year', y='lifeExp', hue='country', kind='line', style='continent', size='continent')
```

```python
# facet plot -> works only with figure level function 
# it will not work with scatterplot and lineplot

sns.relplot(data=tips, x='total_bill', y='tip', row='sex', col='smoker')
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', col='year', col_wrap=3)
```

# 2. Distribution Plots

- Univariate Analysis
- Used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

#### Plots under distribution plot
- histplot
- kdeplot
- rugplot

```python
# Figure Level Function -> displot
# Axes Level Functions -> histplot, kdeplot, rugplot
```

## 2.1. Histplot

```python
# histplot -> plotting univariate histogram
sns.histplot(data=tips, x='total_bill', bins=15)
```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# hue
sns.histplot(data=tips, x='total_bill', hue='sex')
```

```python
# faceting using col and rows
sns.displot(data=tips, x='tip', kind='hist', col='sex')
```

## 2.2. kdeplot

Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate


```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', fill=True, kind='kde')
```

## 2.3. rugplot

- Plot marginal distributions by drawing ticks along the x and y axes.
- This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

## 2.4.1. Bivariate Histogram
 
- A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color

```python
sns.histplot(data=tips, x='total_bill', y='tip')
# sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
```

## 2.4.2. Bivariate Kdeplot

A bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```
