## 5. Apply Function

```python
# split (apply) combine
# apply -> builtin functions
genres.apply(min)
```

```python
# find number of movies starting with A for each group

def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# Using filtering
movies[movies['Series_Title'].str.startswith('A')]['Genre'].value_counts()
```

```python
# find ranking of each movie in the group according to IMDB score
def rank_movie(group):
    group['Genre_Rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating 
def normalize(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normalize)
```

```python
ipl= pd.read_csv(r'C:\Users\SVF\Downloads\Datasets\Grouby/deliveries.csv')
ipl.sample(2)
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs scored
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no. of sixes
six = ipl[ipl['batsman_runs'] == 6]

six.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
six['batsman'].value_counts().idxmax()
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
f1 = ipl['over'] > 15
f2 = (ipl['batsman_runs'] == 4) | (ipl['batsman_runs'] == 6)

temp_df = ipl[f1 & f2]

temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).idxmax()
```

```python
# find v kohli's record against all teams
kohli = ipl[ipl['batsman'] == 'V Kohli']

kohli.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# create a function that can return the highest runs scored by batsman
temp_df = ipl[ipl['batsman'] == 'MS Dhoni']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('V Kohli')
```

```python
highest('SK Raina')
```
