## 5. Selecting Rows from a DataFrame

- iloc-: Searches using index positions
- loc-: Searches using index labels

```python
 student_dict = {
    'name': ['abc', 'mno', 'pqr', 'xyz'],
    'iq':[100, 90, 120, 80],
    'marks':[80, 70, 100, 50],
    'package':[10,7,14,2]
}

students = pd.DataFrame(student_dict)
students.set_index('name', inplace=True)
students
```

```python
# single row
movies.iloc[0]
```

```python
students.loc['abc']
```

```python
# multiple rows
movies.iloc[0:5]
```

```python
students.loc['abc':'xyz']
```

```python
# fancy indexing
movies.iloc[[0,4,5]]
```

```python
students.loc[['abc', 'pqr', 'xyz']]
```

## 7. Filtering a Dataframe

```python
ipl.head(2)
```

```python
# find all the final winners
final = ipl[ipl['MatchNumber'] == 'Final']
final[['Season', 'WinningTeam']]
```

```python
# how many super over finishes have occures
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches csk won in kolkata
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage
((ipl['TossWinner'] == ipl['WinningTeam']).sum() / ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and votes > 10000
movies.head(2)
```

```python
movies[(movies['imdb_rating'] >= 8) & (movies['imdb_votes'] > 10000)]
```

```python
# Action movies with rating higher than 7.5

#action = movies['genres'].str.split('|').apply(lambda x: 'Action' in x)
action = movies['genres'].str.contains('Action')
rating = movies['imdb_rating'] > 7.5

movies[action & rating]
```
