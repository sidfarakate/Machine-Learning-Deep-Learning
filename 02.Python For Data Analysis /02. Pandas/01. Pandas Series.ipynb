## Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Q-9

i. Read `items.csv` making `item_name` as index. <br>
ii. Show no of nan values <br>
ii. Item price is given in $, so convert it to rupees without currency symbol. <br>
iii. Make data type of newly made series as float. <br>
iv. Fill nan with mean of the series

```python
# Read items.csv making item_name as index.
df = pd.read_csv(r'C:\Users\SVF\Downloads\Datasets\Pandas_Series\items.csv', index_col='item_name').squeeze()
df.head(2)
```

```python
#  Show no of nan values
df.isna().sum()
```

```python
# Item price is given in $, so convert it to rupees without currency symbol.
def rupees(x):
    try:
        y = x[1:]
    except:
        # Sometimes Dollar sign is not there which throws an exception 
        y = x
    return float(y)*83.50
```

```python
items = df.apply(rupees)
items
```

```python
#  Make data type of newly made series as float.
df = pd.to_numeric(items)
```

```python
# Fill nan with mean of the series
df.fillna(df.mean())
```
