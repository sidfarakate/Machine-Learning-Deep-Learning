# 1. 2D Line Plot

- Used For Bivariate Analysis
- Categorical -> Numerical and Numerical -> Numerical
- Use Case -> Time Series Data

image.png

```python
# plotting a simple function
price = [45000, 49000, 57000, 51000, 48000, 65000] # Numerical Feature
year = [2019, 2020, 2021, 2022, 2023, 2024] # Categorical Feature

plt.plot(year, price)
```

```python
# from a pandas dataframe
df = pd.read_csv(r'C:\Users\SVF\Downloads\Datasets\Plotting Using Matplotlib\sharma-kohli.csv')
df.sample(2)
```

```python
plt.plot(df['index'], df['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(df['index'], df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])
```

```python
# labels and title
plt.plot(df['index'], df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
# hex codes: https://htmlcolorcodes.com/
# linestyle -> (solid, dashed, dotted, dashdot)
# marker -> (D, +, ., >, <, o, x)

plt.plot(df['index'], df['V Kohli'], color='red', linewidth=3, marker='x', markersize=10)
plt.plot(df['index'], df['RG Sharma'], color='blue', linestyle='dashdot', linewidth=2, marker='D')

plt.title('Rohit Sharma Vs Virat Kohli Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend
plt.plot(df['index'], df['V Kohli'], color='red', linewidth=3, marker='x', markersize=10, label = 'V Kohli')
plt.plot(df['index'], df['RG Sharma'], color='blue', linestyle='dashdot', linewidth=2, marker='D', label= 'R Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# loc -> ('best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 
#         'lower center', 'upper center', 'center')
plt.legend(loc='lower right')
```

```python
# limiting axes
# Because of the single outlier, curve looks like a flat
price = [45000, 49000, 57000, 51000, 48000, 650000] 
year = [2019, 2020, 2021, 2022, 2023, 2024] 
plt.plot(year, price)
```

```python
price = [45000, 49000, 57000, 51000, 48000, 65000] # Numerical Feature
year = [2019, 2020, 2021, 2022, 2023, 2024] # Categorical Feature

plt.plot(year, price)
plt.ylim(40000, 60000)
# plt.xlim(2021, 2023)
```

```python
# grid -> for improving the readability
plt.plot(df['index'], df['V Kohli'], color='red', linewidth=3, marker='x', markersize=10, label = 'V Kohli')
plt.plot(df['index'], df['RG Sharma'], color='blue', linestyle='dashdot', linewidth=2, marker='D', label= 'R Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.grid()
```

# 3. Bar Chart

- Bivariate Analysis
- Numerical vs Categorical
- Use Case - Aggregate Analysis Of Groups


image.png

```python
# simple bar chart
children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color='yellow')
plt.xlabel('Colors')
plt.ylabel('Number_Childrens')
```

```python
# horizontal bar chart
plt.barh(colors, children, color='red')
```

```python
df = pd.read_csv(r'C:\Users\SVF\Downloads\Datasets\Plotting Using Matplotlib\batsman_season_record.csv')
df
```

```python
# multiple bar chart
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2, label=2015)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, label=2016)
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2, label=2017)

# plt.xticks(existing labels, new labels)
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.legend()
```

```python
# problem -> x labels overlapped
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors, children)
```

```python
plt.bar(colors, children, color='green')
plt.xticks(rotation='vertical')

plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015'] + df['2016']),label='2017')

plt.legend()
plt.grid()
plt.show()
```

# 5. Pie Chart

- Univariate/Bivariate Analysis
- Categorical vs Numerical
- Use Case - To Find Contribution On A Standard Scale


image.png

```python
# simple data
data = [23, 45, 100, 20, 49]
subjects = ['English', 'Science', 'Maths', 'History', 'Hindi']

plt.pie(data, labels=subjects)
plt.show()
```

```python
df = pd.read_csv(r'C:\Users\SVF\Downloads\Datasets\Plotting Using Matplotlib\gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
