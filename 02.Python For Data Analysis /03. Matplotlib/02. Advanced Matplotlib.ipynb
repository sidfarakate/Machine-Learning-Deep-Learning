# 4. 3D Plots

## 4.1 3D Scatter Plot

```python
fig = plt.figure()

ax = plt.subplot(projection='3d')
ax.scatter3D(df['runs'], df['avg'], df['strike_rate'], marker='+')

ax.set_title('IPL Batsman Analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
```

## 4.2. 3D Line Plot

```python
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

fig = plt.figure()

ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z, s=[100, 100, 100, 100])
ax.plot3D(x,y,z, color='red')
```

## 4.3. 3D Surface Plot

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x,y)

z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## 4.4. Contour Plots

```python
z = xx**2 + yy**2

fig = plt.figure(figsize=(8,6))

ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(8,6))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
